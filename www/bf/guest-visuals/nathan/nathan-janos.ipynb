```python
df = pd.read_csv('item_daily-totals_AdvertisingAttributionwithNathanJanos_2015-07-04.csv')
df['tick'] = df['day'] == 1
df.head()
```

```python
plt.figure(figsize=(10,3))
ticks = df['tick']
x = np.arange(df.shape[0])
plt.plot(x, df.total_downloads, color='#6da251')
xx = df[ticks].index
yy = df[ticks]['date']
plt.xticks(df[ticks].index, df[ticks]['date'].tolist(), rotation=90)
plt.ylabel('Downloads', fontsize=16)
plt.title('Downloads per day from episode 3 featuring Nathan Janos', fontsize=16)
plt.gca().yaxis.grid(True)
plt.savefig('nathan.png',bbox_inches='tight')
plt.show()
```
