```python
%matplotlib inline
import matplotlib.pyplot as plt
import requests
import pandas as pd
import numpy as np
```

```python
first = 1970
years = np.arange(first, 2016)
```

```python
sightings = np.zeros(len(years))
sightings[0] = 1
sightings[len(sightings)-1]=1
```

```python
fig = plt.figure(figsize=(10,3))
plt.bar(years, sightings, color='#6da251')
plt.ylim(0, 1.1)
plt.ylabel('Confirmations', fontsize=16)
plt.title('Confirmed scientifically documented\nsightings of Bouvier''s Red Colobus Monkey', fontsize=16)
filter = years % 5 == 0
plt.yticks([0,1], [0,1])
plt.xlim(1970, 2016)
plt.xticks(years[filter] + 0.5, years[filter], rotation=90)
plt.gca().annotate('Lieven\'s expedition confirm\nspecies not extinct', xy=(2015, .8),
                xytext=(-300, -80), textcoords='offset points',
                arrowprops=dict(arrowstyle="->",
                                connectionstyle="arc,angleA=0,armA=30,rad=10"), fontsize=16
                )

plt.savefig('lieven.png',bbox_inches='tight')
plt.show()
```
