```python
%matplotlib inline
import matplotlib.pyplot as plt
import requests
import pandas as pd
import numpy as np
```

```python
years = np.arange(1920, 2016)
```

```python
sightings = np.zeros(len(years))
sightings[0] = 1
sightings[len(sightings)-1]=1
```

```python
1970
2015
```

```python
fig = plt.figure(figsize=(10,3))
plt.bar(years, sightings, color='#6da251')
plt.ylim(0, 2)
plt.ylabel('Confirmations', fontsize=16)
plt.title('Confirmed scientifically documented\nsightings of the Red Colobus Monkey', fontsize=16)
filter = years % 5 == 0
plt.xticks(years[filter] + 0.5, years[filter], rotation=90)
plt.gca().annotate('Lieven''s photograph''s confirm species still exists', xy=(2015, 1),
                xytext=(-300, 30), textcoords='offset points',
                arrowprops=dict(arrowstyle="->",
                                connectionstyle="arc,angleA=0,armA=30,rad=10"),
                )

plt.savefig('lieven.png',bbox_inches='tight')
plt.show()
```
